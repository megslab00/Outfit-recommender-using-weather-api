```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt

# Load Fashion MNIST
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

# Preprocess
x_train = x_train / 255.0
x_test = x_test / 255.0

# Add channel dimension (for CNN)
x_train = x_train[..., tf.newaxis]
x_test = x_test[..., tf.newaxis]

# Build a simple CNN
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D(),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')  # 10 clothing categories
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train
history=model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))

# Evaluate
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"Test accuracy: {test_acc:.2f}")
```

```python


# Plot training and validation accuracy
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import requests
import pandas as pd
from IPython.display import Image, display, HTML

# Open-Meteo Weather API Integration
def get_weather(city):
    # First get latitude/longitude from city name
    geo_url = f"https://geocoding-api.open-meteo.com/v1/search?name={city}&count=1"
    geo_response = requests.get(geo_url).json()
    
    if not geo_response.get('results'):
        return {"error": "City not found"}
    
    lat = geo_response['results'][0]['latitude']
    lon = geo_response['results'][0]['longitude']
    
    # Get current weather
    weather_url = f"https://api.open-meteo.com/v1/forecast?latitude={lat}&longitude={lon}&current_weather=true"
    weather_response = requests.get(weather_url).json()
    
    temp = weather_response['current_weather']['temperature']
    weather_code = weather_response['current_weather']['weathercode']
    
    # Map weather code to conditions
    weather_conditions = {
        0: "clear", 1: "clear", 2: "cloudy", 3: "cloudy",
        45: "fog", 48: "fog",
        51: "drizzle", 53: "drizzle", 55: "drizzle",
        56: "freezing drizzle", 57: "freezing drizzle",
        61: "rain", 63: "rain", 65: "rain",
        66: "freezing rain", 67: "freezing rain",
        71: "snow", 73: "snow", 75: "snow",
        77: "snow grains",
        80: "rain showers", 81: "rain showers", 82: "rain showers",
        85: "snow showers", 86: "snow showers",
        95: "thunderstorm", 96: "thunderstorm", 99: "thunderstorm"
    }
    
    return {
        "temperature": temp,
        "conditions": weather_conditions.get(weather_code, "clear"),
        "weather_code": weather_code
    }

# Outfit Recommendation Function (Updated)
def recommend_outfit(city, df):
    weather_data = get_weather(city)
    
    if "error" in weather_data:
        return weather_data["error"]
    
    temp = weather_data["temperature"]
    condition = weather_data["conditions"]
    
    # Improved weather mapping
    if temp > 25:
        weather = "sunny"
    elif 15 <= temp <= 25:
        weather = "rainy" if "rain" in condition else "cloudy"
    elif 0 <= temp < 15:
        weather = "rainy" if "rain" in condition else "cold"
    else:
        weather = "cold"
    
    # Get matching outfits
    matching_outfits = df[df["weather"] == weather]
    
    # If no exact match, find closest temperature range
    if len(matching_outfits) == 0:
        temp_diff = df["temperature_range"].str.extract(r'(-?\d+)-(-?\d+)').astype(float)
        avg_temp = (temp_diff[0] + temp_diff[1]) / 2
        closest_idx = (avg_temp - temp).abs().idxmin()
        recommended = df.iloc[closest_idx]
    else:
        recommended = matching_outfits.sample(1).iloc[0]
    
    # Display recommendation
    display(HTML(f"<h2>🌤️ Recommended Outfit for {city} ({temp}°C, {condition})</h2>"))
    display(HTML("<div style='display: flex; justify-content: space-between;'>"))
    display(HTML("<div style='width: 30%; text-align: center;'>"))
    display(HTML("<h3>👕 Top</h3>"))
    display(Image(url=recommended["top_image"], width=200))
    display(HTML(f"<p>{recommended['top']}</p>"))
    display(HTML("</div>"))
    
    display(HTML("<div style='width: 30%; text-align: center;'>"))
    display(HTML("<h3>👖 Pants</h3>"))
    display(Image(url=recommended["pants_image"], width=200))
    display(HTML(f"<p>{recommended['pants']}</p>"))
    display(HTML("</div>"))
    
    display(HTML("<div style='width: 30%; text-align: center;'>"))
    display(HTML("<h3>👟 Footwear</h3>"))
    display(Image(url=recommended["footwear_image"], width=200))
    display(HTML(f"<p>{recommended['footwear']}</p>"))
    display(HTML("</div>"))
    display(HTML("</div>"))

# Sample Dataset
outfit_data = {
    "weather": ["sunny", "sunny", "rainy", "rainy", "cold", "cold"],
    "temperature_range": ["25-35", "20-25", "10-20", "5-10", "0-5", "-5-0"],
    "top": ["T-shirt", "Tank top", "Rain jacket", "Sweater", "Puffer jacket", "Thermal top"],
    "top_image": [
        "https://images.unsplash.com/photo-1529374255404-311a2a4f1fd9",
        "https://images.unsplash.com/photo-1576566588028-4147f3842f27",
        "https://images.unsplash.com/photo-1539533018447-63fcce2678e4",
        "https://images.pexels.com/photos/974911/pexels-photo-974911.jpeg",
        "https://images.unsplash.com/photo-1551488831-00ddcb6c6bd3",
        "https://images.pexels.com/photos/1183266/pexels-photo-1183266.jpeg"
    ],
    "pants": ["Shorts", "Jeans", "Waterproof pants", "Jeans", "Thermal leggings", "Snow pants"],
    "pants_image": [
        "https://images.unsplash.com/photo-1598033129183-c4f50c736f10",
        "https://images.unsplash.com/photo-1473966968600-fa801b869a1a",
        "https://images.unsplash.com/photo-1591047139829-d91aecb6caea",
        "https://images.pexels.com/photos/1598507/pexels-photo-1598507.jpeg",
        "https://images.pexels.com/photos/1183266/pexels-photo-1183266.jpeg",
        "https://images.unsplash.com/photo-1515886657613-9f3515b0c78f"
    ],
    "footwear": ["Sandals", "Sneakers", "Rain boots", "Boots", "Winter boots", "Snow boots"],
    "footwear_image": [
        "https://images.unsplash.com/photo-1543163521-1bf539c55dd2",
        "https://images.unsplash.com/photo-1460353581641-37baddab0fa2",
        "https://images.pexels.com/photos-298863/pexels-photo-298863.jpeg",
        "https://images.unsplash.com/photo-1549298916-f52d724204b4",
        "https://images.pexels.com/photos/1456706/pexels-photo-1456706.jpeg",
        "https://images.unsplash.com/photo-1515886657613-9f3515b0c78f"
    ]
}

df = pd.DataFrame(outfit_data)

# Example usage
recommend_outfit("Tokyo", df)  # Try different cities like "Paris", "New York", etc.
```
